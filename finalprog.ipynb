```python
import random
import re
import nltk
import webbrowser
user="USER:"
priyanka="BOT: "
quit="bye"
responses={"thank you":["you are welcome"],"hi":["hi"],"hello":["Hi"],"how are you ?":["I am fine","Thanks for asking, I am good","Good. And you?"],"what is your name ?":["I am BOT","Its BOT here","My name is BOT"],"how old are you?":["I'm 20","20"],"what's your age?":["I'm 20","20"],"are you single":["Well that's a personal question","No personal questions please"],"i love you":["Aww! Good to know that","Really?"],"yes":["Great!","Okay"],"no":["Fine","Okay"],"okay":["Yes",";)"]}
new=2
tabUrl="http://www.google.com/?#q="
def talk_to_me(message):
    message=message.lower()
    print(message)
    while message not in quit:
        response=respond(message)
        print(priyanka)
        print(response)
        print(user)
        message=input()
        message=message.lower()
def respond(message):
    if message in responses:
        return random.choice(responses[message])
    else:
        response=matching(message)
        return response
def matching(message):
    pattern1={"do","you","?"}
    pattern2={"what","who","is","are","?"}
    token=nltk.word_tokenize(message)
    domatch=re.compile("do you .*")
    domatch2=re.compile("you are .*")
    domatch3=re.compile("what is .*")
    domatch4=re.compile("i .*")
    domatch5=re.compile("can you .*")
    domatch6=re.compile("who is .*")
    pos=nltk.pos_tag(token)
    if domatch.match(message):
        reply="Yes I "
        i=0
        for w in token:
            if w not in pattern1:
                if 'PRP' in pos[i]:
                    w=replace_pronoun(w)
                reply=reply+" "+w
            i=i+1
        return reply
    elif domatch2.match(message):
        return "Thanks!"
    elif domatch5.match(message):
        return "yes"
    elif domatch4.match(message):
        return "That's great !"
    elif domatch3.match(message) or domatch6.match(message):
        reply=""
        for w in token:
            if w not in pattern2:
                reply=reply+" "+w
        webbrowser.open(tabUrl+reply,new=new)
        return reply
    else:
        return "OK"
def replace_pronoun(reply):
    replace=reply.lower()
    if 'me' in reply:
        return re.sub('me','you',reply)
    if 'my' in reply:
        return re.sub('my','your',reply)
    if 'your' in reply:
        return re.sub('your','my',reply)
    if 'you' in reply:
        return re.sub('you','I',reply)
    return reply
```
